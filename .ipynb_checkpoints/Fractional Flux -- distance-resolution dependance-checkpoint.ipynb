# How stable is Fractional Flux?

Is this value,
$$ \text{fractional flux} = \frac{\sum_{\text{counts in pixel}}^{\text{counts at SN}} \text{counts in pixel}}{\sum_{\text{all pixels}} \text{counts in pixel}} $$
and its distribution in a "complete" data set, stable when viewing of galaxies over differenant distances or with telescopes of different angular resolution.

First off, I think that it is a combination of distance ($d$) and pixel size ($\theta$). So really it is the physical resolution of the observed object, or $d\times\theta = s$.
